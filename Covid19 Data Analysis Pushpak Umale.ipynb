**COVID-19**


    The coronavirus outbreak came to light on December 31, 2019 when China informed the World Health Organisation of a cluster of cases of pneumonia of an unknown cause in Wuhan City in Hubei Province. Subsequently the disease spread to more Provinces in China, and to the rest of the world. The WHO declared it a pandemic. The virus has been named SARS-CoV-2 and the disease is now called COVID-19 (source : "The Hindu")

    In this Dataset, I've analysed data regarding the Covid19 pandemic downloaded from Kaggle. The Data used in this project is recorded as on 29 April 2020 (i,e during the initial phases of Covid in India and the World)

**The Things that I've analysed in this model are :**

    1. The Total number of Confirmed, Death and Recovered people from the dataset
    2. Top 20 countries with the most confirmed Cases
    3. Top 50 countires with the most and least deaths
    4. Top 60 countries with the most recovery 
    5. India's Status as of 29th April 2020
    6. USA's Status as of 29th April 2020
    6. Sorting with respect to Recovery rate


```python
#importing necessary libraries
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as mplt
```

```python
os.getcwd() #finding the current working directory in order to save the dataset in it
```

```python
covdata = pd.read_csv('/home/jovyan/binder/covid_19_data.csv')
```

```python
covdata
```

```python
covdata.count()
```

```python
covdata.isnull().sum()
```

```python
sns.heatmap(covdata.isnull())#to see the non non-values results in heatmap form
mplt.show()
```

```python
#to show the total number of confirmed, deaths and recovered cases in each region 
#1. confirmed
covdata.groupby('Region')['Confirmed'].sum()
```

```python
#2. deaths
covdata.groupby('Region')['Deaths'].sum()
```

```python
#3. recovered
covdata.groupby('Region')['Recovered'].sum()
```

```python
#top 20 regions with the most confirmed cases
covdata.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(20)
```

```python
#Top 60 regions with the most number of deaths
covdata.groupby('Region')['Deaths'].sum().sort_values(ascending = False).head(60)
```

```python
# Top 60 regions with the least number of deaths
covdata.groupby('Region')['Deaths'].sum().sort_values(ascending = True).head(60)
```

```python
## Top 60 regions with the highest recovery
covdata.groupby('Region')['Recovered'].sum().sort_values(ascending = False).head(60)
```

```python
#reporting India's status as on 29th april 2020
covdata[covdata.Region == 'India']
```

```python
#reporting USA's status as on 29th april 2020
covdata[covdata.Region == 'US']
```

```python
#sorting the entire data with respect to the number of recovered cases (top 20)
#1. Ascending
covdata.sort_values(by=['Recovered'], ascending = True).head(20)
```

```python
#2. Descending
covdata.sort_values(by=['Recovered'], ascending = False).head(20)
```
